```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# Fungsi untuk membaca gambar dan melakukan operasi opening

def proses_opening(nama_file):
    if not os.path.exists(nama_file):
        print(f"File '{nama_file}' tidak ditemukan.")
        return None, None

    image = cv2.imread(nama_file, cv2.IMREAD_GRAYSCALE)
    _, binary = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    kernel = np.ones((5, 5), np.uint8)
    opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel)
    return binary, opening

# Proses gambar pertama
img1, hasil1 = proses_opening('A.jpeg')
# Proses gambar kedua
img2, hasil2 = proses_opening('B.jpeg')

plt.figure(figsize=(12, 6))

if img1 is not None:
    plt.subplot(2, 2, 1)
    plt.title("A.jpeg - Asli")
    plt.imshow(img1, cmap='gray')
    plt.axis('off')

    plt.subplot(2, 2, 2)
    plt.title("A.jpeg - Opening")
    plt.imshow(hasil1, cmap='gray')
    plt.axis('off')

if img2 is not None:
    plt.subplot(2, 2, 3)
    plt.title("B.jpeg - Asli")
    plt.imshow(img2, cmap='gray')
    plt.axis('off')

    plt.subplot(2, 2, 4)
    plt.title("B.jpeg - Opening")
    plt.imshow(hasil2, cmap='gray')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
